Exception is used to raised the error when any error has occured in program.
If any error occur then without using eception error has been occured but rest all the statement after that get not executed because of that error so by exception we can continue our program even any error get occur.

```python
# when exception is not handled then because of one error rest all the statement will not get execute.
f=open("text2.txt",'r')
print("this statement ill not get executed")
```

```python
# except keyword is used to handled the error
try:
    f=open("text2.txt",'r')
except Exception as e:
    print(e)
print("my statement will execute")
```

```python
# try and else
try:
    f=open("text.txt",'r')
except Exception as e:
    print(e)
else:
    print("try block successfully get executed")

#finally
try:
    f=open("text1.txt",'r')
    f.write("this is file")
    f.close()
except Exception as e:
    print(e) #not run as the try statement has not produced any error
else:
    print("try block successfully get executed")
finally:
    print("this will always executed in any condition") # this will always executed in any condition

#raise
class raising_error(Exception):
    def __init__(self,msg):
        self.msg=msg
def val_age(age):
    if age<0:
        raise raising_error("negative age is not valid")
    elif age>200:
        raise raising_error("age is very large")
    else:
        print("valid age")
try:
    age=int(input("inter the age"))
    val_age(age)
except raising_error as e:
    print(e)
```
